```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()#transform cata to num
for col in dataset.columns:
  if dataset[col].dtype=='object':#current type  is catagorical value
    dataset[col]=le.fit_transform(dataset[col])#to transform the num
```

```python
imputer=SimpleImputer(strategy='median')#used for filli the missing value
mis_cols=['RiskLevel']
dataset[mis_cols]=imputer.fit_transform(dataset[mis_cols])#replaing the missing value
```

```python
dataset.head()
```

#standardition feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=dataset.iloc[:,:6]
y=dataset['RiskLevel']
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
x=pd.DataFrame(x)
```

```python
x.head()
```

#model building

```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
logreg = lr(max_iter=1000)
```

```python
logreg.fit(x_train, y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```
